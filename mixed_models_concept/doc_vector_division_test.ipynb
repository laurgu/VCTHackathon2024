```python
import os
import pandas as pd
import json
```

```python
# Load datasets needed
ch_players = pd.read_json("Cleaned-Jsons/ch_players.json")[['handle', 'team_name', 'region', 'league_name']]
gc_players = pd.read_json("Cleaned-Jsons/gc_players.json")[['handle', 'team_name', 'region', 'league_name']]
intl_players = pd.read_json("Cleaned-Jsons/intl_players.json")[['handle', 'team_name', 'region', 'league_name']]

player_stats = pd.read_json("Cleaned-Jsons/allplayer_stats.json")
player_stats
```

```python
# Rename column for player handle in player stats
ch_players_filtered = ch_players.rename(columns={'handle': 'Player'})
ch_players_filtered
```

```python
# Rename column for player handle in player stats
gc_players_filtered = gc_players.rename(columns={'handle': 'Player'})
gc_players_filtered
```

```python
# Rename column for player handle in player stats
intl_players_filtered = intl_players.rename(columns={'handle': 'Player'})
intl_players_filtered
```

```python
# Merge all player stats per league
merged_ch = pd.merge(player_stats, ch_players_filtered, on='Player', how='left')
merged_ch_filtered = merged_ch.dropna(subset=['region', 'league_name']) # Filter by active and have team. Assume those without region are not in an active league
merged_ch_filtered
```

```python
# Merge all player stats per league
merged_gc = pd.merge(player_stats, gc_players_filtered, on='Player', how='left')
merged_gc_filtered = merged_gc.dropna(subset=['region', 'league_name']) # Filter by active and have team. Assume those without region are not in an active league
merged_gc_filtered
```

```python
# Merge all player stats per league
merged_intl = pd.merge(player_stats, gc_players_filtered, on='Player', how='left')
merged_intl_filtered = merged_intl.dropna(subset=['region', 'league_name']) # Filter by active and have team. Assume those without region are not in an active league
merged_intl_filtered
```

```python
# Export results
merged_ch_filtered.to_parquet('player_stats_splits/league_split/challengers_players_stats.parquet')
merged_gc_filtered.to_parquet('player_stats_splits/league_split/gamechangers_players_stats.parquet')
merged_intl_filtered.to_parquet('player_stats_splits/league_split/international_players_stats.parquet')
```

```python
# Combined players
merged_all = pd.concat([merged_ch_filtered, merged_gc_filtered, merged_intl_filtered], ignore_index=True).reset_index(drop=True)
merged_all
```

```python
merged_all.to_parquet('player_stats_splits/league_split/all_leagues_players_stats.parquet')
```
